## Format column names 
Same features have different column names across different datasets

```python
current_feature = list(df_current.columns.values)
```

```python
## change all column names to lowercases 
for i in range (len(current_feature)):
    current_feature[i] = str.lower(str(current_feature[i]))
```

```python
current_feature[:5]
```

```python
## remove '_' in column names
his_feature = list(df_historical.columns.values)
for i in range (len(his_feature)):
    his_feature[i] = str(his_feature[i]).replace('_','')
```

```python
his_feature[:5]
```

## Find common features between current and historical data

```python
common_feature = set(his_feature)&(set(current_feature))
print('numer of common features: ', len(common_feature))
print('common features:', common_feature)
```

```python
current_unique = set(current_feature)-set(his_feature)
print('number of features unique to current data: {} \n {} \n'.format(len(current_unique), current_unique))
```

```python
his_unique = set(his_feature)-set(current_feature)
print('number of features unique to historical data: {} \n {} \n'.format(len(his_unique), his_unique))
```

## Create feature dictionary

```python
feature_dic={}
feature_dic['common_feature']=common_feature
feature_dic['his_unique']=his_unique
feature_dic['current_unique']=current_unique
```

```python
## add common features that have different names for current and historical data
## Use the form in the historical data)

common_feature = common_feature.union({'zipcode', 
 'fundedamnt',
 'verificationstatus',
 'verificationstatusjoint',
 'loanamnt',
 'numacctsever120pd'})

print('number of common features: ', len(common_feature))
```

```python
df_historical = df_historical[list(common_feature)+['issued', 'loanstatus']]
```
